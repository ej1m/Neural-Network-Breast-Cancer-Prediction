# Seperating features and targets

```python
Y = data_frame['label']
Y
```

```python
X = data_frame.drop(columns = 'label' , axis = 1)
X
```

# Normalize x features

```python
from sklearn.preprocessing import StandardScaler
#create a scaler object
scaler = StandardScaler()
#fit scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# split in training and test sets
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size = 0.3, random_state = 2)
print(X.shape, x_train.shape, x_test.shape)
```

# Build neural network

```python
import tensorflow as tf  # for developing deep learning models (neural networks)
# keras used tensorflow as its backend to make it easier to develop neural networks
tf.random.set_seed(3) # prevents accuracy score from changing regardless of how many times you run code
from tensorflow import keras
```

```python
# set up neural network layer structure
#dense - all neurons in a layer are connected to all neurons in the previous layer
model = keras.Sequential([
# input layer, flatten - x_train is a matrix, when training neural networks, we have to convert to 1-d array
    keras.layers.Flatten(input_shape=(30,)), 
# hidden layer, 20 neurons
    keras.layers.Dense(20, activation = 'relu'), 
# output layer, here, the number of neurons in output layer = no of classes(0,1) = 2
    keras.layers.Dense(2, activation = 'sigmoid' ) 
]) 
```

```python
model.compile(optimizer='adam',
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# training neural network
# validation set- 10% of data
# goal - reduce loss value, increase accuracy value
# epochs- no of time model goes throgh data
history = model.fit(x_train, y_train,validation_split=0.1, epochs = 30) 
```

```python
# visualizing model accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
# visualizing model loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc = 'upper right')
```

# Model accuracy on test data

```python
# Evaluate model loss and accuracy on test data
loss, accuracy = model.evaluate(x_test, y_test)
print (f'model loss is: {loss:.2f}\n model accuracy is:{accuracy*100:.2f}%')
```

# Building the predictive system

```python
def model_prediction (input_data):
    # change input data from tuple to numPy array
    input_data_as_numPy_array = np.asarray(input_data)
    # reshape numpy array as we're predicting label for one data point
    input_data_reshaped = input_data_as_numPy_array.reshape(1,-1)
    # standardize input_data
    input_data_std = scaler.transform(input_data_reshaped)
    pred = model.predict(input_data_std)
    print(pred)
    pred_label = [np.argmax(pred)]
    print(pred_label)
    if pred_label[0] == 0:
        print('the tumor is malignant')
    else:
        print('the tumor is benign')
    return pred_label
```
